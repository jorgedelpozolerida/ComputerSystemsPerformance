```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re

PYTORCH_DIR = os.path.join(".", "gpu", "pytorch")
```

```python
def parse_data(dir_path):
    df_list = []
    
    for path in os.listdir(dir_path):
        if os.path.isfile(os.path.join(dir_path, path)) and path.endswith("ENERGY.csv"):
            file_parts = re.split('_|-', path)
            sub_df = pd.read_csv(os.path.join(dir_path, path))
            sub_df["run"] = file_parts[1]
            sub_df["device"] = file_parts[3]
            sub_df["batch_size"] = file_parts[7]
            sub_df["framework"] = file_parts[9]
            sub_df["dataset"] = file_parts[11]
            
            df_list.append(sub_df)
    return pd.concat(df_list).reset_index()
            
parse_data(PYTORCH_DIR)
```
