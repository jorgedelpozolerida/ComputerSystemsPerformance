```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
concurrect_df = pd.concat([
    pd.read_csv(f"../experiments/concurrent_output/experiment_{x}.csv", sep=";")
    for x in range(1,9)]).reset_index(drop=True)

concurrect_df["Running Time (s)"] = concurrect_df["Running Time (ms)"]/1000
concurrect_df["total tuples"] = 2000000
concurrect_df["million tuples per second"] = concurrect_df["total tuples"]*concurrect_df["Running Time (s)"]
concurrent_df = concurrect_df.drop(columns=["Running Time (ms)", "Running Time (s)", "total tuples"]).groupby(['Threads', 'Hash_Bits']).mean()
concurrect_df
```

```python
sb.lineplot(data=concurrect_df, x="Hash_Bits", y="million tuples per second", hue="Threads")
```
