```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
concurrect_df = pd.concat([pd.read_csv(f"../experiments/concurrent_output/experiment_{1}.csv", sep=";")]).reset_index(drop=True)

concurrect_df["total tuples"] = 2000000
concurrect_df["tuples per ms"] = concurrect_df["total tuples"].div(concurrect_df["Running Time (ms)"].values)
concurrect_df
```

```python
sb.lineplot(data=concurrect_df, x="Hash_Bits", y="tuples per ms", hue="Threads").set(title="Concurrent Output")
```

```python
ind_df = pd.concat([pd.read_csv(f"../experiments/independent_output/experiment_{1}.csv", sep=";")]).reset_index(drop=True)

ind_df["total tuples"] = 2000000
ind_df["tuples per ms"] = ind_df["total tuples"].div(ind_df["Running Time (ms)"].values)
ind_df
```

```python
sb.lineplot(data=ind_df, x="Hash_Bits", y="tuples per ms", hue="Threads").set(title="Independent Output")
```
